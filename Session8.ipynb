```python
try:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    c = a / b
    print("c =", c)

except ZeroDivisionError:
    print("You are trying to divide a number by 0")
```

# index error

```python
lst = [1, 2, 3, 4, 5, 6, 7, 10]
print(lst[7])   # last element
```

```python
try:
    lst = [10, 20, 33, 44, 12, 33, 44, 45]
    indexnumber = int(input("Enter your index number: "))
    print(lst[indexnumber])

except IndexError:
    print("List index out of range")
```

# type error

```python
a = "shdusdf"
b = 23
c = a + str(b)
print(c)
```

```python
try:
    a = 'shdusdf'
    b = 23
    c = a + b
    print(c)

except TypeError as e:
    print(e)
```

# rise keyword

```python
try:
    age = int(input("Enter your age: "))
    if age < 18:
        raise TypeError("Person is not eligible for voting")
    else:
        print("Person is eligible for voting")
except TypeError as e:
    print(e)
```

```python
try:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    c = a + b
    print("c =", c)

except ValueError as e:
    print("Invalid input! Please enter numeric values only.")

finally:
    print("Finally your calculation is completed")
```
