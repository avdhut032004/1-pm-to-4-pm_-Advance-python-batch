# Operator Overloading

```python
# + = __add__
# - = __sub__
# * = __mul__
# / = __div__

# & = __and__
# | = __or__

# > = __gt__
# < = __lt__
# > = = __gte__
# < = = __lte__
# = = __eq__
```

```python
class Addition:
    def __init__(self,a):
        self.a=a
    def __add__(self,other):
        return self.a+other.a 
```

```python
obj1=Addition(20)
obj2=Addition(20)
```

```python
obj1+obj2
```

```python
class compare_age:
    def __init__(self):
        self.name=input("Enter your name")
        self.age=int(input("Enter your age"))
    def __gt__(self,other):
        if self .age>other.age:
            return f"{self.name} is older than {other.name}"
        else:
            return f"{self.name} is younger than {other.name}"    
```

```python
person2=compare_age()
```

```python
person1=compare_age()
```

```python
person1 > person2
```

```python
class check_divisibility:
    def __init__(self):
        self.num=int(input("Enter a Number"))
    def __and__(self,other):
        if self.num%2==0 and other.num%2==0:
            return "both number is even"
        else:
            return "Both number are odd"
```

```python
o1=check_divisibility()
```

```python
o2=check_divisibility()
```

```python
def check_divisibility():
    num = int(input("Enter a number: "))
    if num % 2 == 0:
        print(num, "is divisible by 2")
    else:
        print(num, "is not divisible by 2")
    return num

o1 = check_divisibility()
o2 = check_divisibility()

addition = o1 + o2
print("Addition of both numbers:", addition)
```

```python
# complex number = combination of real and imaginary number
# a+ib
# a=10,b=20
# a=4,b=10
# res=14+i10
```

```python
class complex_number:
    def __init__(self,a,b):
        self.a=a
        self.b=b
    def __add__(self,other):
        return f"{self.a+other.a}+i{self.b+other.b}"    
```

```python
c1=complex_number(4,7)
```

```python
c2=complex_number(9,3)
```

```python
c1+c2 
```
